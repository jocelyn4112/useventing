```python
# remove Texas Rose Horse Park Fall HT
df1=df1[df1['Event'] != 'Texas Rose Horse Park Fall HT']
df1.index
```

```python
# remove rows with no dressage score
df1=df1[df1['Dress Score'] != ' ']
df1.index
```

```python
# remove Virgina International & H.T.(May 2019) CCI Two Star - L Division
df1=df1[((df1['Event'] != 'Virginia International & H.T.(May 2019)') & (df1['Division'] != 'CCI Two Star - L'))]
df1.index
```

```python
# identifies and corrects rows with misaligned column data
n=0
for index, row in df1.iterrows():
    check=row['Stad Time']
    if ':' in check:
        newxcj=row['Stad Jump']
        newxce=row['Stad Time']
        newxct=row['To Date Score']
        newfs=row['To Date Place']
        df1.loc[index, ['Stad Jump', 'Stad Time', 'To Date Score', 'To Date Place']]=[row['XC Jump'], row['XC Elapsed'], row['XC Time'], row['Final Score']]
        df1.loc[index, ['XC Jump', 'XC Elapsed', 'XC Time', 'Final Score']]=[newxcj, newxce, newxct, newfs]
        n=n+1
        
n
```

```python
#Examines and notes index of nonsense data
n=0
droprows = []
for index, row in df1.iterrows():
    temp3 = row['Stad Jump']
    if '-T' in temp3:
        n=n+1
        temp=row['Event']
        temp2=row['Division']
        temp4=row['Rider']
        print(f"Event: {temp}, Div: {temp2}, Stad Time:  {row['Stad Time']}")
        droprows.append(index)
n
```

```python
# drops nonsense data
df1=df1.drop(droprows, axis=0)
df1.index
```

```python
n=0
checklist=['Dress Score', 'Stad Jump', 'Stad Time', 'To Date Score', 'XC Jump',
       'XC Elapsed', 'XC Time', 'Final Score']
for index, row in df1.iterrows():
    for check in checklist:
        if '-T' in row[check]:
            n=n+1
            print(f"-T found: {index}, {check}, {row[check]}")
n

```

```python
df1['XCE']=np.nan
df1['xcs']=df1['XC Elapsed'].str.split(pat=':', n=1)
for index, row in df1.iterrows():
    x=(int(row.xcs[0])* 60)+(int(row.xcs[1]))
    df1.loc[index, 'XCE'] = x
df1.head(50)
# df1[['XC Elapsed', 'b4', 'aftr', 'XC Elapsedsec']].head(50)
```

```python
#creates a list of columns that contain fall data and a test dataset that contains only fall rows
falls=[]
fallindexes=[]
for index, row in df1.iterrows():
    for column in df1.columns:
        if row[column] == 'RF':
            if column not in falls:
                falls.append(column)
            if index not in fallindexes:
                fallindexes.append(index)
falldf=df1[df1.index.isin(fallindexes)]
falldf.head(50)
```

```python
df1.sfall=(df1['Stad Jump'] =='RF')
#            | (df1['Stad Time'] == 'RF'))
# .astype('int')
df1.sfall.value_counts()
```

```python
new=[]
for stuff in falls:
    varname=stuff+'fall'
    new.append(varname)
    falldf[varname]=(falldf[stuff]=='RF').astype('int')
falldf.fall = 0
for things in new:
    for index, row in falldf.iterrows():
        if row[things]:
            falldf.loc[index, 'fall']=1
```
