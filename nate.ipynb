```python
# remove Texas Rose Horse Park Fall HT
df1=df1[df1['Event'] != 'Texas Rose Horse Park Fall HT']
df1.index
```

```python
# remove rows with no dressage score
cutlist=[' ', 'TE', '0', 'E', 'R', 'W']
df1=df1[~df1['Dress Score'].isin(cutlist)]
df1.index
```

```python
# remove Virgina International & H.T.(May 2019) CCI Two Star - L Division
df1=df1[((df1['Event'] != 'Virginia International & H.T.(May 2019)') & (df1['Division'] != 'CCI Two Star - L'))]
df1.index
```

```python
# identifies and corrects rows with misaligned column data
n=0
for index, row in df1.iterrows():
    check=row['Stad Time']
    if ':' in check:
        newxcj=row['Stad Jump']
        newxce=row['Stad Time']
        newxct=row['To Date Score']
        newfs=row['To Date Place']
        df1.loc[index, ['Stad Jump', 'Stad Time', 'To Date Score', 'To Date Place']]=[row['XC Jump'], row['XC Elapsed'], row['XC Time'], row['Final Score']]
        df1.loc[index, ['XC Jump', 'XC Elapsed', 'XC Time', 'Final Score']]=[newxcj, newxce, newxct, newfs]
        n=n+1
        
n
```

```python
#Examines and notes index of nonsense data
n=0
droprows = []
for index, row in df1.iterrows():
    temp3 = row['Stad Jump']
    if '-T' in temp3:
        n=n+1
        temp=row['Event']
        temp2=row['Division']
        temp4=row['Rider']
        print(f"Event: {temp}, Div: {temp2}, Stad Time:  {row['Stad Time']}")
        droprows.append(index)
n
```

```python
# drops nonsense data
df1=df1.drop(droprows, axis=0)
df1.index
```

```python
# checks for any remaining -T in rows where they should not be
n=0
checklist=['Dress Score', 'Stad Jump', 'Stad Time', 'To Date Score', 'XC Jump',
       'XC Elapsed', 'XC Time', 'Final Score']
for index, row in df1.iterrows():
    for check in checklist:
        if '-T' in row[check]:
            n=n+1
            print(f"-T found: {index}, {check}, {row[check]}")
n
```

```python
#confirms no dressage divisions remain in the dataset
n=0
for index, row in df1.iterrows():
    if 'dressage' in row['Division'].lower():
        print(row['Division'], row['Event'])
        n=n+1
n
```

```python
#assigns new variable (difficulty) based upon key word searches in division
indies=[]
df1['Difficulty']=0
# df1['Horse']=0
n=0
newdic={'beginner novice': 2,
        'novice': 3, 
        'training': 4,
        'preliminary': 5,
        'prelim': 5, 
        'intermediate': 6,
        'advanced': 7}
for index, row in df1.iterrows():
    temp=row['Division'].lower()
    for key in newdic.keys():
            if key == 'beginner novice':
                if ('beginner' in temp) & ('novice' in temp) | ('nov' in temp): 
                    df1.loc[index, 'Difficulty']=newdic[key]
                    if index not in indies:
                        indies.append(index)
                    n+=1
                    if n < 1000:
                        print(f'Division: {temp} encoded as {newdic[key]}')
            elif key == 'novice':
                    if ('beginner' not in temp) & (key in temp):
                        df1.loc[index, 'Difficulty']=newdic[key]
                        if index not in indies:
                            indies.append(index)
                        n+=1
                        if n < 1000:
                            print(f'Division: {temp} encoded as {newdic[key]}')
            elif key in temp.lower():
                    df1.loc[index, 'Difficulty']=newdic[key]
                    if index not in indies:
                        indies.append(index)
                    n+=1
                    if n < 1000:
                        print(f'Division: {temp} encoded as {newdic[key]}')
```

```python
new_df=df1[~df1.index.isin(indies)]
new_df.index
```

```python
# assigns difficulty based on star level amongst rows with difficulty not classified above
n=0
anotherdic={'1': 4, '2': 5, '3':6, '4': 7, 'one':4, 'two': 5, 'three': 6, 'four':7}
for index, row in new_df.iterrows():
    temp=row['Division'].lower()
    if ('cci' in temp) | ('cic' in temp):
        for stuff in anotherdic.keys():
            if stuff in temp:
                if index not in indies:
                        indies.append(index)
                df1.loc[index, 'Difficulty']=anotherdic[stuff]
                n+=1
                if n < 1000:
                    print(f'Division: {temp}: {anotherdic[stuff]}')
            
```

```python
features = np.c_[categorical_Difficulty, X[:,1], X[:,2], categorical_Senior, categorical_Junior, categorical_CT, categorical_Event,
                categorical_Rider_x]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, y, random_state=1)
```

```python
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers.experimental import preprocessing
```

```python
def modelbuilder(hp):
    model = Sequential()
    hp_units=hp.Int('units', min_value=10, max_value=100, step=20)
    model.add(Dense(units=hp_units, activation='relu', input_dim=features.shape[1]))
    model.add(Dense(units=hp_units, activation='relu'))
    model.add(Dense(units=2, activation='softmax'))
    model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
    return model
```

```python
tuner=kt.Hyperband(modelbuilder, objective='val_accuracy', max_epochs=10, factor=3)
stop_early=keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=10)
```

```python
tuner.search(X_train, y_train_categorical, epochs=100, validation_split=0.2, callbacks=[stop_early])
```

```python
# !pip install -U keras-tuner
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]
best_hps
```

```python
model=tuner.hypermodel.build(best_hps)
```

```python
history=model.fit(X_train, y_train_categorical, epochs=100, validation_split=0.2)
```

```python
accuracy=history.history['val_accuracy']
best_epoch=accuracy.index(max(accuracy))
best_epoch
```

```python
hypermodel=tuner.hypermodel.build(best_hps)
hypermodel.fit(
    X_test,
    y_test_categorical,
    epochs=10,
    shuffle=True,
    verbose=2
)
```

```python
hypermodel.evaluate(X_test, y_test_categorical)
```
