```python
X=final_df.drop(['Fall'], axis=1)
# .values
y=final_df['Fall'].values
# print(X.shape, y.shape)
```

```python
X.dtypes
```

```python
X=X.values
```

```python
categorical_Difficulty=to_categorical(X[:, 0])
categorical_Senior=to_categorical(X[:, 3])
categorical_Junior=to_categorical(X[:, 4])
categorical_CT=to_categorical(X[:, 5])
categorical_Event=to_categorical(X[:, 6])
categorical_Rider_x=to_categorical(X[:, 7])
# categorical_Horse_x=to_categorical(X[:, 8])
# features=[]
# for x in np.arange(len(X)):
#     features.append([X[x, 0], X[x, 1], categorical_Event[x], categorical_Division[x], categorical_Rider[x]])    
```

```python
features = np.c_[categorical_Difficulty, X[:,1], X[:,2], categorical_Senior, categorical_Junior, categorical_CT, categorical_Event,
                categorical_Rider_x]
```

```python
# features=np.asarray(features)
features.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, y, random_state=1)
```

```python
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=features.shape[1]))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(
    X_train,
    y_train_categorical,
    epochs=5,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
predictions = model.predict_classes(X_test)
predictions.shape
```

```python
from sklearn.metrics import confusion_matrix
con=confusion_matrix(y_test, predictions)
print(f'True Negatives: {con[0][0]}')
print(f'False Negatives: {con[1][0]}')
print(f'True Positives: {con[1][1]}')
print(f'False Positives: {con[0][1]}')
```
