```python
big_data_merged_weather = pd.merge(
    big_data_df_final,
    weather_sums,
    how="left",
    on=['Event']
    #indicator=True
    #validate="many_to_one"
)

big_data_final =big_data_merged_weather.drop_duplicates()

big_data_final
```

```python
list_var = ['Lon', 
                'Lat',
                'temp-end',
                'precip-end',
                'wspd-end',
                'visibility-end',
                'cloudcover-end',
                'wgust-end',
                'moonphase-end',
                'Precip_Prior_3_Day',
                'Precip_Prior_5_Day',
                'Precip_Prior_7_Day']

n=0

# big_data_df_final[list_var] = big_data_df_final[list_var]'NaN'
for things in list_var:
    big_data_df_final[things] = 'NaN'

for index,row in big_data_df_final.iterrows():
    
    temp = row['Event']
    
    for var in list_var:

        try:    

            big_data_df_final.loc[index,[var]] = weather_index.loc[[temp]][var][0]

        except:
            n+=1
```
