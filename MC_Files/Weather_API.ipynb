```python
import requests
import pandas as pd
```

```python
location_df = pd.read_csv('../MC_Files/venue_location_MC_Clean_2_18_21_11PM.csv')
location_df
```

```python
#extract unique Merge_Candidates aka venues and drop 'Needs Key Lookup' value

location_df1 = location_df.drop_duplicates(subset=['Merge_Candidate'])

location_df1 = location_df1[location_df1['Merge_Candidate'] != 'Needs Key Lookup']

#check drop
location_df1.loc[location_df1['Merge_Candidate'] == 'Needs Key Lookup']

location_df1
```

```python
#read in big dataset with dates 
big_data_df = pd.read_csv('../MC_Files/eedata_dates_MC_Clean_2_18_21_10PM.csv')
big_data_df
```

```python
#extract unique dates from big_data_df

big_data_df1 = big_data_df.drop_duplicates(subset=['Start_Datetime'])
big_data_df1
```

```python
 #merge with location_df1 based on 'Event' and 'Merge Candidate'
joined_df = pd.merge(
    big_data_df1,
    location_df1,
    how="inner",
    left_on=['Event'],
    right_on=['Merge_Candidate'],
)

joined_df.head()

for col in joined_df.columns:
    print(col)
```

```python
#drop extra columns for new df 

api_df = joined_df[['Event', 'Start_Datetime', 'End_Datetime', 
                   'Lon', 'Lat', 'EventVenueAddress', 'EventVenueFacilityName',
                   'EventVenueCity', 'EventVenueState', 'EventVenueZip']]

api_df
```

```python
#add date ranges for weather api 

for index,row in api_df.itterows:
    api_df['Datetime-1'] = api_df['Start_Datetime'] - pd.DateOffset(days=1)
    api_df['Datetime-2'] = api_df['Start_Datetime'] - pd.DateOffset(days=2)
    api_df['Datetime-3'] = api_df['Start_Datetime'] - pd.DateOffset(days=3)
    api_df['Datetime-4'] = api_df['Start_Datetime'] - pd.DateOffset(days=4)
    api_df['Datetime-5'] = api_df['Start_Datetime'] - pd.DateOffset(days=5)
    api_df['Datetime-6'] = api_df['Start_Datetime'] - pd.DateOffset(days=6)
    api_df['Datetime-7'] = api_df['Start_Datetime'] - pd.DateOffset(days=7)
```
