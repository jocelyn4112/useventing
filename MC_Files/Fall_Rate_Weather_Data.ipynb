```python
import pandas as pd
```

```python
fall_df = pd.read_csv('../final_df_v5.csv')
fall_df
```

```python
#find denoted fall column 

fall_df.loc[fall_df['Fall'] == 1].count()
```

```python
fall_df_clean = fall_df.dropna(subset=['Lat', 'Lon'])
```

```python
fall_df_clean
```

```python
fall_df_clean.loc[fall_df_clean['Fall'] == 1]

#739/26203 is ~3% 
```

```python
#groupby event to calculate falls per event 

fall_df_clean['Event_Falls'] = fall_df_clean.groupby('Event')['Fall'].transform('sum')

fall_df_clean.loc[fall_df_clean['Event_Falls'] > 0]
```

```python
#Count number of riders to get total number of rides per event 

fall_df_clean['Ride_Count'] = fall_df_clean.groupby('Event')['Event'].transform('size')
```

```python
#fall rate columns and round decimals

fall_df_clean['Fall_Rate_Decimal'] = fall_df_clean['Event_Falls'] / fall_df_clean['Ride_Count']
fall_df_clean['Fall_Rate_Percent'] = (fall_df_clean['Event_Falls'] / fall_df_clean['Ride_Count']) * 100

fall_df_clean.round(2)
```

```python
map_df = fall_df_clean.drop_duplicates(subset=['Event'])
map_df
```

```python
map_df.to_csv('fall_rate.csv', index=False)
```
