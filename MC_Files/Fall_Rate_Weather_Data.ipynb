```python
import pandas as pd
```

```python
fall_df = pd.read_csv('../final_df_v5.csv')
fall_df
```

```python
#find denoted fall column 

fall_df.loc[fall_df['Fall'] == 1].count()
```

```python
fall_df_clean = fall_df.dropna(subset=['Lat', 'Lon'])
```

```python
fall_df_clean['Event'].unique()
```

```python
fall_df_clean.loc[fall_df_clean['Fall'] == 1]

#739/26203 is ~3% 
```

```python
#groupby event to calculate falls per event 

fall_df_clean['Event_Falls'] = fall_df_clean.groupby('Event')['Fall'].transform('sum')

fall_df_clean.loc[fall_df_clean['Event_Falls'] > 0]
```

```python
#Count number of riders to get total number of rides per event 

fall_df_clean['Ride_Count'] = fall_df_clean.groupby('Event')['Event'].transform('size')
```

```python
#fall rate columns and round decimals

fall_df_clean['Fall_Rate_Decimal'] = fall_df_clean['Event_Falls'] / fall_df_clean['Ride_Count']
fall_df_clean['Fall_Rate_Percent'] = (fall_df_clean['Event_Falls'] / fall_df_clean['Ride_Count']) * 100

fall_df_clean.round(2)
```

```python
fall_df_clean.to_csv('fall_rate_analysis_map.csv', index=False)
```

```python
map_df = fall_df_clean.drop_duplicates(subset=['Event'])
map_df
```

```python
#create smaller map dataset 

map_df_final = map_df[['Event', 'Lon', 'Lat', 'temp-end', 'precip-end', 'wspd-end', 'visibility-end', 'cloudcover-end', 
                      'moonphase-end', 'Precip_Prior_3_Day', 'Precip_Prior_5_Day', 'Precip_Prior_7_Day', 'Event_Falls', 
                      'Ride_Count', 'Fall_Rate_Decimal', 'Fall_Rate_Percent']]

map_df_final
```

```python
#bring in datetime data 
location_data = pd.read_csv('../MC_Files/weather_sums_final.csv')
location_data
```

```python
location_data2 = location_data[['Event', 'End_Datetime']]
location_data2
```

```python
#merge with fall rate dataset to append date

merged_map = pd.merge(
    map_df_final,
    location_data2,
    left_on=['Event'],
    right_on=['Event']
)

merged_map
```

```python
merged_map.to_csv('event_map_dataset.csv', index=False)
```
