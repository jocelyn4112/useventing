```python
X_test_categorical.shape
```

```python
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
def modelbuilder(hp):
    model = Sequential()
    hp_units=hp.Int('units', min_value=10, max_value=100, step=20)
    model.add(Dense(units=hp_units, activation='relu', input_dim=X_train_categorical.shape[1]))
    model.add(Dense(units=hp_units, activation='relu'))
    model.add(Dense(units=2, activation='softmax'))
    model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
    return model
```

```python
tuner=kt.Hyperband(modelbuilder, objective='val_accuracy', max_epochs=10, factor=3)
stop_early=keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=10)
```

```python
tuner.search(X_train_categorical, y_train_categorical, epochs=100, validation_split=0.2, callbacks=[stop_early])
```

```python
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]
best_hps
```

```python
model=tuner.hypermodel.build(best_hps)
```

```python
history=model.fit(X_train_categorical, y_train_categorical, epochs=10, validation_split=0.2)
```

```python
accuracy=history.history['val_accuracy']
best_epoch=accuracy.index(max(accuracy))
best_epoch
```

```python
hypermodel=tuner.hypermodel.build(best_hps)
hypermodel.fit(
    X_train_categorical,
    y_train_categorical,
    epochs=10,
    shuffle=True,
    verbose=1
)
```

```python
hypermodel.evaluate(X_test_categorical, y_test_categorical)
```

```python
predictions = hypermodel.predict_classes(X_test_categorical)
predictions.shape
```
